```python
# ============================================================
# HR ATTRITION ANALYSIS - DATA CLEANING
# Step 1: Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np

# ============================================================
# Step 2: Upload Dataset
# ============================================================

from google.colab import files

uploaded = files.upload()

# ============================================================
# Step 3: Load CSV File
# ============================================================

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

# ============================================================
# Step 4: Display First 5 Records
# ============================================================

df.head()
```

```python
# ============================================================
# Step 5: Dataset Information
# ============================================================

print("Dataset Shape:", df.shape)

print("\n--------------------------------------------------")
print("Column Names:")
print(df.columns)

print("\n--------------------------------------------------")
print("Data Types:")
print(df.dtypes)

print("\n--------------------------------------------------")
print("Missing Values:")
print(df.isnull().sum())

print("\n--------------------------------------------------")
print("Duplicate Records:", df.duplicated().sum())
```

```python
# ============================================================
# Step 6: Statistical Summary
# ============================================================

print("====================================================")
print("Statistical Summary")
print("====================================================")

df.describe()
```

```python
# ============================================================
# Step 7: Check Unique Values
# ============================================================

categorical_columns = df.select_dtypes(include='object').columns

for column in categorical_columns:
    print(f"\n========== {column} ==========")
    print(df[column].unique())
```

```python
# ============================================================
# Step 8: Remove Unnecessary Columns
# ============================================================

df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'], inplace=True)

print("Columns Removed Successfully!")

print("\nNew Dataset Shape:", df.shape)
```

```python
# ============================================================
# Step 9: Export Cleaned Dataset
# ============================================================

output_file = "HR_Attrition_Cleaned.csv"

df.to_csv(output_file, index=False)

print("Cleaned dataset exported successfully!")
print("File Name:", output_file)
```
